```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\ASUS\\Desktop\\Matchine learning\\Data sets\\Bodyfat.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.drop(columns="bodyfat")
y=df["bodyfat"]
```

```python
plt.figure(figsize=(30,8))
sns.pairplot(x_vars=['Density', 'Age', 'Weight', 'Height', 'Neck', 'Chest',
       'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
       'Wrist'],y_vars=["bodyfat"],data=df)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain.head()
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)

```

```python
X_poly=poly_reg.fit_transform(x)
```

```python
X_poly
```

```python
lin_reg.fit(x,y)
```

```python
lin_reg.score(x,y)
```

```python
pol_reg.fit(X_poly,y)
```

```python
pol_reg.score(X_poly,y)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse=mean_squared_error(ytest,ypred) 
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
mae=mean_absolute_error(ytest,ypred)
mae
```

```python
score=r2_score(ytest,ypred)
score
```
